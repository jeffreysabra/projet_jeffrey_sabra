```python
from sklearn.neighbors import NearestNeighbors

def machinelearning(input_title, finalmerge):
    # Créer une liste avec le film à chercher
    movies_to_search = [input_title]

    # Supprimer les films à chercher de la DataFrame
    movie_without = finalmerge.loc[finalmerge['title'] != input_title]

    # Sélectionner les colonnes numériques pour l'entrainement du modèle
    X = movie_without.select_dtypes('number')

    # Entrainement du modèle
    model = NearestNeighbors(n_neighbors=5)
    model.fit(X)

    # Chercher les films à partir desquels on doit recommander
    for name in movies_to_search:
        stats = finalmerge.loc[finalmerge['title'] == name, X.columns]
        result = model.kneighbors(stats)
        distances = result[0][0]
        indices = result[1][0]

    # Obtenir la liste de recommandations triée par ordre croissant de distance
    recommendations = list(movie_without.iloc[indices]['title'])
    distances_sorted, recommendations_sorted = zip(*sorted(zip(distances, recommendations)))

    #Retourner les recommandations triées par ordre croissant de distance
    return recommendations_sorted

# Demander à l'utilisateur de saisir un film à chercher
movie_to_search = input("Entrez un film à chercher : ")

# Obtenir la liste de recommandations de films similaires
recommendations = machinelearning(movie_to_search, finalmerge)

# Afficher le texte sans les crochets
print("Recommandations pour", movie_to_search, ":")
print(recommendations)
```
